```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv("https://raw.githubusercontent.com/tofighi/MachineLearning/master/datasets/student_marks.csv")
```

```python
df.shape
```

```python
learning_rate = 0.0001
m = -0.5
b=0
```

```python
def standardize(x):
    std = (x - np.mean(x)) / np.std(x)
    return std
```

```python
def error(x, y, m, b):
    err = np.mean((y - (m * x + b))**2)
    return err
```

```python
def gradient_m(x, y, m, b):
    grad_m = -2 * np.mean(x * (y - (m * x + b)))
    return grad_m
```

```python
def gradient_b(x, y, m, b):
    grad_b = -2 * np.mean(y - (m * x + b))
    return grad_b
```

```python
def update_weights(x, y, m, b, learning_rate):
    new_m = m - learning_rate * gradient_m(x, y, m, b)
    new_b = b - learning_rate * gradient_b(x, y, m, b)
    return new_m, new_b
```

```python
plt.scatter(df["Midterm mark"],df["Final mark"])
x = np.linspace(df["Midterm mark"].min(), df["Midterm mark"].max(), 100)
y = m * x + b
plt.plot(x, y, 'r')
```

```python
errors1 = []
m=-0.5
b=0
for i in range(100):
    m, b = update_weights(df["Midterm mark"], df["Final mark"], m, b, learning_rate)
    y = m * x + b
    errors1.append(error(df["Midterm mark"], df["Final mark"], m, b))

print(m,b, errors1)
```
